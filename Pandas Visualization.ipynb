Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\ranjb\Downloads\Ice Cream Ratings.csv")
df
```

```python
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('fast')
```

```python
df.plot()
```

```python
df.plot(kind='line', title = 'Ice cream Ratings')
```

```python
df.plot(kind='line', subplots = True, xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
df.plot(kind = 'bar')
```

```python
df.plot(kind = 'bar', stacked = True)
```

```python
df['Texture Rating'].plot(kind = 'bar', stacked = True, color = 'Orange')
```

```python
df.plot.barh(stacked = True)
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Flavor Rating', s = 600, c = 'Pink')
```

```python
df.plot.hist(bins = 25)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10,5))
```

```python
df.plot.pie(y='Texture Rating',figsize=(10,10))
```
